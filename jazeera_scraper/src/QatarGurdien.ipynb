```python
finallist=[]
```

```python
try:
    for y in range (1,50):
        for j in range(0,10):
                    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.84 Safari/537.36'}
                    response = requests.get(info[y]['response']['results'][j]['webUrl'], headers=headers)
                    page_soup = BeautifulSoup(response.content, features="lxml")
                    full_text = ""
                    for x in page_soup.find_all('p'):
                        full_text = full_text + x.get_text() + "\n"
                    response = requests.get(info[y]['response']['leadContent'][j]['webUrl'], headers=headers)
                    page_soup = BeautifulSoup(response.content, features="lxml")
                    full_text_1 = ""
                    for x in page_soup.find_all('p'):
                        full_text_1 = full_text_1 + x.get_text() + "\n"


                    value=dict(webtitle=info[y]['response']['results'][j]['webTitle'],
                    publisheddate=info[y]['response']['results'][j]['webPublicationDate'],Text = full_text)
                    
                    value1=dict(webtitle=info[y]['response']['leadContent'][j]['webTitle'],
                    publisheddate=info[y]['response']['leadContent'][j]['webPublicationDate'],Text = full_text_1)
                    if value not in finallist :  
                        finallist.append(value)
                    if value1 not in finallist :  
                        finallist.append(value1)
except IndexError:
    print("done")
```

```python
unique_list_of_dicts = [dict(t) for t in {tuple(d.items()) for d in finallist}]
```
